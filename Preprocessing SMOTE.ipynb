**<center>Impor Modul</center>**

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from collections import Counter
```

**<center>Impor Data</center>**

```python
df = pd.read_csv('C:/Statistika ITS/TA/Data/data_kanker_payudara_all_stage.csv', sep=';')
df.columns = ['stadium', 'usia', 'riwayat_keluarga', 'status_hormonal', 
              'riwayat_penyakit_payudara', 'ER', 'PR', 'HER2', 'lama_menderita']
df
```

```python
df["stadium"] = df["stadium"].astype("category")
```

**<center>Variabel Dummy</center>**

```python
categorical_val = []
continous_val = []
for column in df.columns:
    print('==============================')
    print(f"{column} : {df[column].unique()}")
    if len(df[column].unique()) <= 10:
        categorical_val.append(column)
    else:
        continous_val.append(column)
```

```python
categorical_val.remove('stadium')
dataset = pd.get_dummies(df, columns = categorical_val)
dataset
```

**<center>Standardisasi</center>**

```python
s_sc = StandardScaler()
col_to_scale = ['usia', 'lama_menderita']
dataset[col_to_scale] = s_sc.fit_transform(dataset[col_to_scale])
dataset
```

```python
dataset.info()
```

**<center>SMOTE</center>**

```python
X = dataset.drop('stadium', axis=1)
y = dataset.stadium
```

```python
counter = Counter(y)
print(counter)
```

```python
# transform the dataset
oversample = SMOTE(random_state=0, sampling_strategy = {1: 112, 4: 130}, k_neighbors=3)
X, y = oversample.fit_resample(X, y)
counter = Counter(y)
print(counter)
```

```python
result = pd.concat([X, y], axis=1)
result
```

```python
writer = pd.ExcelWriter('data_after_SMOTE.xlsx')
result.to_excel(writer)
writer.save()
```
